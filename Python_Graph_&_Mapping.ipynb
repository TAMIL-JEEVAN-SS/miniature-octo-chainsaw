```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("Python Graph & Mapping/Beverage.csv")

corr = df.corr(numeric_only=True)

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap - Beverage Dataset")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Python Graph & Mapping/FastFood1.csv")

avg_time = df.groupby("Delivery")["Time"].mean()

avg_time.plot(kind="bar", figsize=(10,5))
plt.ylabel("Average Delivery Time")
plt.title("Average Delivery Time by Delivery Type")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("Python Graph & Mapping/rating.csv")

# Convert wide format to long format
df_melted = df.melt(
    var_name="Channel",
    value_name="Rating"
)

plt.figure(figsize=(8,5))
sns.violinplot(x="Channel", y="Rating", data=df_melted)
plt.title("Rating Distribution by Channel")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Python Graph & Mapping/Florida.csv")

# Select only the date columns (exclude Metropolitan Area)
population_data = df.drop(columns=["Metropolitan Area"])

# Convert columns to time axis, specifying the format
population_data.columns = pd.to_datetime(population_data.columns, format='%b_%Y')

# Sum population across all metro areas for each time period
total_population = population_data.sum()

plt.figure(figsize=(8,5))
plt.plot(total_population.index, total_population.values, marker='o')
plt.xlabel("Time")
plt.ylabel("Population")
plt.title("Florida Population Trend")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Python Graph & Mapping/Melbourne_Housing_NoMissing.csv")

features = df.groupby("Type")[["Rooms", "Bedroom", "Bathroom", "Car"]].mean()

features.plot(kind="bar", stacked=True, figsize=(10,5))
plt.ylabel("Average Count")
plt.title("Average Property Features by Property Type")
plt.legend(title="Features")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("Python Graph & Mapping/Melbourne_Housing_NoOutliers.csv")

plt.figure(figsize=(8,5))
sns.histplot(df["Price"], kde=True)
plt.title("Price Distribution (No Outliers)")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Python Graph & Mapping/Melbourne_Housing.csv")

pivot = df.pivot_table(
    values="Price",
    index="Regionname",
    columns="Type",
    aggfunc="mean"
)

plt.figure(figsize=(10,6))
sns.heatmap(pivot, cmap="YlGnBu", annot=True, fmt=".0f")
plt.title("Average House Price by Region & Type")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Python Graph & Mapping/AOVData.csv")

# Display basic info
print(df.head())
print(df.info())

manufacturer_avg = df.groupby("manufacturer")["co_emissions"].mean()

plt.figure()
plt.bar(manufacturer_avg.index, manufacturer_avg.values)
plt.xlabel("Manufacturer")
plt.ylabel("Average CO Emissions")
plt.title("Average CO Emissions by Manufacturer")
plt.show()

fuel_avg = df.groupby("fuel_type")["co_emissions"].mean()

plt.figure()
plt.bar(fuel_avg.index, fuel_avg.values)
plt.xlabel("Fuel Type")
plt.ylabel("Average CO Emissions")
plt.title("Average CO Emissions by Fuel Type")
plt.show()

plt.figure()
plt.hist(df["co_emissions"], bins=10)
plt.xlabel("CO Emissions")
plt.ylabel("Frequency")
plt.title("Distribution of CO Emissions")
plt.show()

plt.figure()
```
